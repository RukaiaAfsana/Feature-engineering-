### replace a list by another list

```python
students = {
    'name': ['afsana', 'promi','safin','adal'],
    'scores' : ['excellent','joss','extraordinary','marhaba']
}
```

```python
df_s=pd.DataFrame(students)
```

```python
df_s
```

```python
df_s.replace({
    'excellent':5.00,
    'joss': 4.00,
    'extraordinary':4.00,
    'marhaba':4.00
})
```

### or

```python
df_s.replace(['excellent','joss','extraordinary','marhaba'],[5.00,4.00,3.00,2.00])
```

## Group By 

```python
df_cities = pd.read_csv("groupby_weather.csv")
```

```python
df_cities
```

```python
g= df_cities.groupby("cities")
```

```python
g
```

```python
for city,city_df in g:
    print(city)
    print(city_df)
```

### find maxium temperature in each of the cities

```python
g.get_group("mumbai").max()
```

```python
g.get_group("new york").max()
```

```python
g.get_group("paris").max()
```

### OR (split,apply,combile)

```python
g.max()
```
